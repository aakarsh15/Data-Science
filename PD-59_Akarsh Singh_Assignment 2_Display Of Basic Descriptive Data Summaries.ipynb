***Display Of Basic Descriptive Data Summaries***

**1) Read and Visualize Total Profit for All Months(Line Plot)**

```python
from google.colab import files
uploaded=files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data_sample.csv", encoding="ISO-8859-1")

total_profit = df.groupby('MONTH_ID')['SALES'].sum()

plt.figure(figsize=(10, 6))
plt.plot(total_profit.index, total_profit.values, marker='o', linestyle='-')
plt.title("Total Profit for All Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
```

**2) Get total profit of all months and show line plot with the following Style properties.**

```python
plt.figure(figsize=(10, 6))
plt.plot(total_profit.index, total_profit.values, marker='o', linestyle='--', color='b', linewidth=2, markersize=8)
plt.title("Total Profit for All Months (Styled)")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
```

**Read all product sales data and show it  using a multiline plot**

```python
plt.figure(figsize=(12, 6))
for product in df['PRODUCTLINE'].unique():  # Loop through unique product categories

    product_data = df[df['PRODUCTLINE'] == product]

    x_values = product_data['MONTH_ID']

    plt.plot(x_values, product_data['SALES'], label=product)

plt.title("Product Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

**Read Classic Cars sales data of each month and show it using a scatter plot**

```python
# Read Classic Cars sales data for each month
classic_cars_sales = df[df['PRODUCTLINE'] == 'Classic Cars']

# Create a scatter plot for Classic Cars sales
plt.figure(figsize=(10, 6))
plt.scatter(classic_cars_sales['MONTH_ID'], classic_cars_sales['SALES'], marker='o', color='b', label='Classic Cars')
plt.title("Classic Cars Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

**Read Trucks and Buses and Classic Cars sales data and show it using the bar chart**

```python
selected_products = ['Trucks and Buses', 'Classic Cars']
selected_sales_data = df[df['PRODUCTLINE'].isin(selected_products)]

plt.figure(figsize=(10, 6))
plt.bar(selected_sales_data['PRODUCTLINE'], selected_sales_data['SALES'])
plt.title("Trucks, Buses, and Classic Cars Sales Data")
plt.xlabel("Product Line")
plt.ylabel("Sales")
plt.grid(axis='y')
plt.show()
```

**Read Total Profit Histogram**

```python
total_monthly_profit = df.groupby('MONTH_ID')['SALES'].sum() - df.groupby('MONTH_ID')['MSRP'].sum()

plt.figure(figsize=(10, 6))
plt.hist(total_monthly_profit, bins=15, edgecolor='k')
plt.title("Total Profit Histogram")
plt.xlabel("Total Profit")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

**Calculate Total Sale Data for Last Year for Each Product in PIE CHART**

```python
last_year_data = df[df['YEAR_ID'] == df['YEAR_ID'].max()]

product_sales_last_year = last_year_data.groupby('PRODUCTLINE')['SALES'].sum()

plt.figure(figsize=(8, 8))
plt.pie(product_sales_last_year, labels=product_sales_last_year.index, autopct='%1.1f%%', startangle=140)
plt.title("Total Sale Data for Last Year by Product")
plt.axis('equal')
plt.show()
```
